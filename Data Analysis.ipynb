```python
#another where the data is from the csv file
import pandas as pd

column = ['Student #', 'Last Name', 'First Name', 'Gender', 'Section', 'Absences', 'Quiz 1', 'Quiz 2', 'Act 1', 'Act 2', 'Act 3', 'Final Exam']

student_df = pd.read_csv('student_data.csv', names = column) 
#pd.read_csv (to read the csv file instead of import path from os)

male = student_df[(student_df['Gender'] == 'M') & (student_df['Section'] == 'A')][['Last Name', 'First Name', 'Section', 'Gender']]

female = student_df[(student_df['Gender'] == 'F') & (student_df['Section'] == 'A')][['Last Name', 'First Name', 'Section', 'Gender']]

print(male)
print('')
print(female)
```
