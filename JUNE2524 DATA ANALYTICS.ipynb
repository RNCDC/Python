```python
#Data cleaning and transformation
#adding and removing columns and rows
import pandas as pd
data = {'Brgy': ['B1','B2','B3','B4'], 
        'Population': [2470,2479,1016,724], 
        'Size': [52,43,30,25]}
brgy_df = pd.DataFrame(data, index = [1,2,3,4])
brgy_df
```

```python
#Single column, same values
#brgy_df[new_col] = value
brgy_df['Hospital'] = None
```

```python
#Adding Columns
#Single column, value from list
#df[new_col] = [value1, value2, value n]
brgy_df['Hospital'] = [2,3,1,0]
```

```python
#Adding Columns at a specific location
#df.insert(position, column, value)
brgy_df.insert(2, 'Voters', 0)
```

```python
#Adding Multiple Columns
#Single Value
#df[[column1, column2, column n]] = None
brgy_df[['Lying-in', 'Evac Area']] = None
```

```python
#Adding Multiple Columns and inserting datas
#From list
#df[[column1, column2, column n]] = None
brgy_df[['Lying-in', 'Evac Area']] = [[5,3], [3,2], [2,1], [1,1]]
brgy_df
```

```python
#Adding Rows
#df.append(data, ignore_index = bool)
add_data = pd.DataFrame([['B5',4000, 70, 6, 5, 3, 1]], columns = brgy_df.columns)
brgy_df = pd.concat([brgy_df, add_data], ignore_index = True)
brgy_df
```

```python
#Removing rows and columns
#df.drop(labels, axis = n, inplace = bool)
#labels - col names or indices to be dropped
#single label - single col/row drop
#lists = multiple col/row drop axis
#0 or 'index'
#1 or 'columns'
#inplace - commit the changes
#bool
```

```python
#Removing columns
#Remove the Lying-in and Evacuation area column. Do not commit changes
temp_df = brgy_df.copy()
temp_df.drop(['Lying-in' , 'Evac Area'], axis = 1)
```

```python
#Removing rows
#remove rows 0 and 4
temp_df.drop([0, 4], axis = 0)
```

```python
#Handling missing data
#data.isnull()  -  Detecting missing values
#data.notnull() -  Detecting non-missing values
#data.isnull().sum()  -  Counting missing values
#data.notnull().sum()  -  Counting non-missing values
#data.dropna(axis = 0, how = 'any', thresh = None, subset = None, inplace = False)  -  Removing data with missing values
#data.fillna(value, inplace = False)  -  Filling-in missing values with some values
```

```python
import pandas as pd
sample = [[2,4.0,70.0,52.0],
         [10,16.0,14.0], 
         [7],
         [5,25.0]]
rows = [1,2,3,4]
col = ['A', 'B', 'C', 'D']
samp2_df = pd.DataFrame(sample, rows, col)
samp2_df
```

```python
#detecting missing values
samp2_df.isnull()
```

```python
#detecting non-missing values
samp2_df.notnull()
```

```python
#dropping data with missing values
#drop rows with any missing value
samp2_df.dropna() #dropna aalisin na
```

```python
#dropping data with missing values
#drop columns with at least 3 missing values
samp2_df.dropna(thresh = 2) #dropna aalisin na
#thresh takes integer value which tells minimum amount of na values to drop
```

```python
#dropping data with missing values
#drop columns with any missing values
samp2_df.dropna(axis = 1) #dropna aalisin na
#Axis specify the dimension of the dataframe in which we want to perform the function in
```

```python
#data.fillna(value, inplace = bool)
samp2_df.fillna(30) #fillna lalagyan laman
```

```python
#more specific ung pag fill in
samp2_df.fillna({'B':5, 'C':10, 'D':20})
```

```python
# Binning data: cut
#When you need to segment and sort data values into bins. 
#This funtion is also useful for going from a continuous variable to categorical variable.
#For example, cut could convert ages to group of age ranges.
#Supports binning into an equal number of bins, or a pre - specified arrays of bins.

# alias.cut(data_array, bins, labels = None)
```

```python
import pandas as pd
data = {'': ['2k-1', '2k-2', '2k-3', '2k-4', '2k-5', '2k-6', '2k-7', '2k-8'], 
        'Names': ['Naruto', 'Sasuke', 'Kakashi', 'Gaara', 'Temari', 'Jiralya', 'Tsunade', 'Itachi'], 
        'TG': [80,90,97,85,93,76,91,65]}
binning_df = pd.DataFrame(data)
binning_df
```

```python
bins = [0,75,80,85,90,93,100]
label = [3,2.5,2,1.75,1.5,1]
binning_df['EQ'] = pd.cut(binning_df['TG'], bins, labels = label)
binning_df
#0-75   3.0
#76-80  2.5
#81-85  2.0
#86-90  1.75
#91-93  1.5
#94-100 1.0
```

```python
#CHATGPT
import pandas as pd

data = {
    '': ['2k-1', '2k-2', '2k-3', '2k-4', '2k-5', '2k-6', '2k-7', '2k-8'], 
    'Names': ['Naruto', 'Sasuke', 'Kakashi', 'Gaara', 'Temari', 'Jiraiya', 'Tsunade', 'Itachi'], 
    'TG': [80, 90, 97, 85, 93, 76, 91, 65]
}
binning_df = pd.DataFrame(data)

bins = [0, 75, 80, 85, 90, 93, 100]
labels = [3, 2.5, 2, 1.75, 1.5, 1]

binning_df['EQ'] = pd.cut(binning_df['TG'], bins=bins, labels=labels, right = False) #right is not included in example
#right = Bool  -  This makes the bins left-inclusive, which means the left edge is 
#included in the bin and the right edge is not.

print(binning_df)
```
