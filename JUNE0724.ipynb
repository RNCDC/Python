```python
from tkinter import*
import tkinter.messagebox as tm
my_window=Tk()
my_window.title('checkbox demo')


chk50=IntVar()
chk60=IntVar()
txtTotal=StringVar()

def on_chk():
    global total
    p50=chk50.get()
    p60=chk60.get()
    if p50==1:
        t50=50
    else:
        t50=0
    if p60==1:
        t60=60
    else:
        t60=0
    total=t50+t60
    #txttotal.set(total)
    
def on_total():
    txtTotal.set(total)
    

Checkbutton(my_window,text='VM 1 P50',variable=chk50,command=on_chk).grid(row=0,column=0)
Checkbutton(my_window,text='VM 2 P60',variable=chk60,command=on_chk).grid(row=0,column=1)
Entry(my_window,textvariable=txtTotal).grid(row=1,column=0)
Button(my_window,text="Total",command=on_total).grid(row=2,column=0)
#width para lumaki box ng text area

my_window.mainloop()
```

```python
#simple calculator number 1 elimination
from tkinter import*
import tkinter.messagebox as tm
my_window=Tk()
my_window.title('Simple Calculator')

#declarations
one=StringVar()
two=StringVar()
txtResult=StringVar()

#method
#to add the values
def on_add():
    try:
        first = float(one.get())
        second = float(two.get())
        total = first + second 
        txtResult.set(total)
    except Exception as z:
        tm.showerror('Error: ',z)

#to subtract the values
def on_minus():
    try:
        first = float(one.get())
        second = float(two.get())
        total = first - second 
        txtResult.set(total)
    except Exception as z:
        tm.showerror('Error: ',z)

#to multiply the values
def on_multiply():
    try:
        first = float(one.get())
        second = float(two.get())
        total = first * second 
        txtResult.set(total)
    except Exception as z:
        tm.showerror('Error: ',z)

#to divide the values
def on_divide():
    try:
        first = float(one.get())
        second = float(two.get())
        total = first / second 
        txtResult.set('%.2f' %total)
    except Exception as z:
        tm.showerror('Error: ',z)

#to clear inputs & outputs
def on_clear():
    one.set('')
    two.set('')
    txtResult.set('')

#layout
Label(my_window,text="First Number").grid(row=0,column=0)
Entry(my_window,textvariable=one).grid(row=0,column=1)
Label(my_window,text="Second Number").grid(row=1,column=0)
Entry(my_window,textvariable=two).grid(row=1,column=1)
Label(my_window,text="Result").grid(row=3,column=0)
Entry(my_window,textvariable=txtResult, state= 'disabled').grid(row=3,column=1)
Button(my_window,text="+",command=on_add).grid(row=4,column=0)
Button(my_window,text="-",command=on_minus).grid(row=4,column=1)
Button(my_window,text="Clear",command=on_clear).grid(row=4,column=2)
Button(my_window,text="*",command=on_multiply).grid(row=5,column=0)
Button(my_window,text="/",command=on_divide).grid(row=5,column=1)
Button(my_window,text="Exit",command=my_window.destroy).grid(row=5,column=2)

my_window.mainloop()
```

```python
#3
from tkinter import*
import tkinter.messagebox as tm
my_window=Tk()
my_window.title('Main Screen')

txtScore=StringVar()
txtItems=StringVar()

def on_next():
    global tg
    screen2=Toplevel(my_window)
    s=float(txtScore.get())
    i=float(txtItems.get())
    
    txtS=StringVar()
    txtI=StringVar()
    txtTG=StringVar()
    txtEQ=StringVar()
    
    tg = s * 50 / i + 50
    
    txtS.set(s)
    txtI.set(i)
    txtTG.set('%.2f'%tg)
    
    Label(screen2,text="Score").grid(row=0,column=0)
    Entry(screen2,textvariable=txtS, state = 'disabled').grid(row=0,column=1)
    Label(screen2,text="Number of Items").grid(row=1,column=0)
    Entry(screen2,textvariable=txtI, state = 'disabled').grid(row=1,column=1)
    Label(screen2,text="Transmuted Grade").grid(row=2,column=0)
    Entry(screen2,textvariable=txtTG, state = 'disabled').grid(row=2,column=1)
    Button(screen2,text="Back",command=screen2.destroy).grid(row=3,column=0)
    Label(screen2,text="EQ").grid(row=3,column=1)
    Entry(screen2,textvariable=txtEQ, state = 'disabled').grid(row=3,column=2)
    
    if tg <= 100 and tg >= 97:
        txtEQ.set('1.00')
    elif tg <= 96 and tg >= 94:
        txtEQ.set('1.25')
    elif tg <= 93 and tg >= 91:
        txtEQ.set('1.50')
    elif tg <= 90 and tg >= 88:
        txtEQ.set('1.75')
    elif tg <= 87 and tg >= 85:
        txtEQ.set('2.00')
    elif tg <= 84 and tg >= 82:
        txtEQ.set('2.25')
    elif tg <= 81 and tg >= 79:
        txtEQ.set('2.50')
    elif tg <= 78 and tg >= 76:
        txtEQ.set('2.75')
    elif tg == 75:
        txtEQ.set('3.00')
    elif tg == 74:
        txtEQ.set('3.25')
    elif tg <= 73 and tg >= 72:
        txtEQ.set('3.50')
    elif tg <= 71 and tg >= 70:
        txtEQ.set('3.75')
    elif tg <= 69 and tg >= 65:
        txtEQ.set('4.00')
    elif tg <= 64 and tg >= 60:
        txtEQ.set('4.25')
    elif tg <= 59 and tg >= 58:
        txtEQ.set('4.50')
    elif tg <= 57 and tg >= 50:
        txtEQ.set('5.00')
    else:
        txtEQ.set("It is Invalid")
    

Label(my_window,text="Score").grid(row=0,column=0)
Entry(my_window,width=20,border=5,textvariable=txtScore).grid(row=0,column=1)
Label(my_window,text="Number of Items").grid(row=1,column=0)
Entry(my_window,width=20,border=5,textvariable=txtItems).grid(row=1,column=1)
Button(my_window,text="Next",command=on_next).grid(row=2,column=0)

my_window.mainloop()
```
