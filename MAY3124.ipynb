```python
import csv
from os import path
from tkinter import *
import tkinter.messagebox as tm
my_window = Tk()
my_window.title('Problem 6')

#methods
#TO CREATING OR APPENDING CSV FILE 
def on_save():
    try:
        if path.exists("ActProb6.csv"): #TO CHECK IF THE FILE IS EXISTING
                fp = open("ActProb6.csv", "a", newline='') #TO APPEND DATA IN CSV FILE
                fprint = csv.writer(fp)
                base = float(s1.get())
                height = float(s2.get())
                side3 = float(s3.get())
                area = 0.5 * base * height
                txtArea.set('%.2f' %area)
                perimeter = base + height + side3
                txtPerimeter.set('%.2f' %perimeter)
                out_list = [base, height, side3]
                fprint.writerow(out_list)
        else:
                fp = open("ActProb6.csv", "w", newline='') #TO CREATE CSV FILE
                fprint = csv.writer(fp)
                fprint.writerow(['Side 1', 'Side 2', 'Side 3'])
                base = float(s1.get())
                height = float(s2.get())
                side3 = float(s3.get())
                area = 0.5 * base * height
                txtArea.set('%.2f' %area)
                perimeter = base + height + side3
                txtPerimeter.set('%.2f' %perimeter)
                out_list = [base, height, side3]
                fprint.writerow(out_list)
        fp.close()
    except Exception as z:
        tm.showerror('Error: ',z)

#TO CLEAR TEXTBOXES (INPUTS AND OUTPUT, DATA IN CSV NOT INCLUDED)
def on_clear():
    s1.set(' ')
    s2.set(' ')
    s3.set(' ')
    txtArea.set(' ')
    txtPerimeter.set(' ')

#TO SHOW THE DATA INSIDE THE CSV FILE ALONG WITH THE COMPUTED AREA AND PERIMETER
def on_view():
    with open('ActProb6.csv','r') as fp:
            fscanf=csv.reader(fp)
            next(fscanf)
            index = 1
            print('\n# \tSide 1 \tSide 2 \tSide 3 \tArea \tPerimeter')
            for i in fscanf:
                z = float(i[0]) 
                y = float(i[1])
                x = float(i[2])
                print("{:<2}\t".format(index),end='')
                index = index + 1
                print('%.2f\t' %z,end='')
                print('%.2f\t' %y,end='')
                print('%.2f\t' %x,end='')
                ar = 0.5 * float(z) * float(y)
                print('%.2f\t' %ar ,end='')
                per = float(z) + float(y) + float(x)
                print('%.2f' %per)
    fp.close()
    
#defining components
s1 = StringVar()
s2 = StringVar()
s3 = StringVar()
txtArea = StringVar()
txtPerimeter = StringVar()

#FOR THE LAYOUT WINDOW
Label(my_window, text = 'Side 1 (Base): ').grid(row = 0, column = 0)
Entry(my_window, textvariable = s1, width = 20, border = 3).grid(row = 0, column = 1)
Label(my_window, text = 'Side 2 (Height): ').grid(row = 1, column = 0)
Entry(my_window, textvariable = s2, width = 20, border = 3).grid(row = 1, column = 1)
Label(my_window, text = 'Side 3: ').grid(row = 2, column = 0)
Entry(my_window, textvariable = s3, width = 20, border = 3).grid(row = 2, column = 1)
Label(my_window, text = 'Area: ').grid(row = 3, column = 0)
Entry(my_window, textvariable = txtArea, width = 20, border = 3, state = 'disabled').grid(row = 3, column = 1)
Label(my_window, text = 'Perimeter: ').grid(row = 4, column = 0)
Entry(my_window, textvariable = txtPerimeter, width = 20, border = 3, state = 'disabled').grid(row = 4, column = 1)
Button(my_window, text = 'Save', command = on_save, width = 10).grid(row = 5, column = 0)
Button(my_window, text = 'Clear', command = on_clear, width = 10).grid(row = 5, column = 1)
Button(my_window, text = 'View', command = on_view, width = 10).grid(row = 5, column = 2)

my_window.mainloop()
```
