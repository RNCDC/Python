```python
#FIXED
from tkinter import *
import tkinter.messagebox as tm

my_window = Tk()
my_window.title('ACTIVITY_PROBLEM #3')

# DECLARATION OF VARIABLES
txtTotal = StringVar()
txtUnits = StringVar()
chkSIS = IntVar()
chkCAT = IntVar()
chkID = IntVar()
chkREG = IntVar()
chkLIB = IntVar()
chkMISC = IntVar()
StudentType = IntVar()

# Initialize global fee variables
vsis = 0
vcat = 0
vid = 0
vreg = 0
vlib = 0
vmisc = 0
STFee = 0

# METHODS
# METHOD FOR WHEN A CHECKBOX IS CLICKED
def on_chk():
    global vsis, vcat, vid, vreg, vlib, vmisc
    
    vsis = 250 if chkSIS.get() == 1 else 0
    vcat = 100 if chkCAT.get() == 1 else 0
    vid = 100 if chkID.get() == 1 else 0
    vreg = 200 if chkREG.get() == 1 else 0
    vlib = 100 if chkLIB.get() == 1 else 0
    vmisc = 100 if chkMISC.get() == 1 else 0

# METHOD FOR WHEN RADIO BUTTONS ARE CLICKED
def on_click():
    global STFee
    ST = StudentType.get()
    if ST == 1:
        STFee = 1
    elif ST == 2:
        STFee = 0.5
    else:
        STFee = 0

# METHOD FOR COMPUTE
def on_compute():
    try:
        units = float(txtUnits.get())
        if StudentType.get() == 0:
            raise ValueError("No student type selected.")
        fees = vsis + vcat + vid + vreg + vlib + vmisc
        tf = (units * 12) + fees
        total = tf - (STFee * tf)
        txtTotal.set('%.2f' % total)
    except ValueError as ve:
        tm.showerror('Error:', ve)
    except Exception as z:
        tm.showerror('Error:', z)

# METHOD FOR CLEAR
def on_clear():
    txtUnits.set('')
    chkSIS.set(0)
    chkCAT.set(0)
    chkID.set(0)
    chkREG.set(0)
    chkLIB.set(0)
    chkMISC.set(0)
    StudentType.set(0)
    txtTotal.set('')
    on_chk()  # Reset the fee variables
    on_click()  # Reset the student type fee

# INPUT FOR UNITS
Label(my_window, text='UNITS ENROLLED').grid(sticky='w', row=0, column=0)
Entry(my_window, textvariable=txtUnits, width=20, border=3).grid(row=0, column=1, sticky='w')

# CHECK BUTTONS FOR OTHER FEES
Label(my_window, text='OTHER FEES').grid(row=2, column=0, sticky='w')
Checkbutton(my_window, text='SIS P250', variable=chkSIS, command=on_chk).grid(row=3, column=0, sticky='w')
Checkbutton(my_window, text='CATALYST 100', variable=chkCAT, command=on_chk).grid(row=3, column=1, sticky='w')
Checkbutton(my_window, text='ID 100', variable=chkID, command=on_chk).grid(row=4, column=0, sticky='w')
Checkbutton(my_window, text='REGISTRATION 200', variable=chkREG, command=on_chk).grid(row=4, column=1, sticky='w')
Checkbutton(my_window, text='LIBRARY 100', variable=chkLIB, command=on_chk).grid(row=5, column=0, sticky='w')
Checkbutton(my_window, text='MISC 100', variable=chkMISC, command=on_chk).grid(row=5, column=1, sticky='w')

# RADIO BUTTON FOR STUDENT TYPE
Label(my_window, text='STUDENT TYPE').grid(row=7, column=0, sticky='w')
Radiobutton(my_window, text='FULL SCHOLAR', value=1, variable=StudentType, command=on_click).grid(row=8, column=0, sticky='w')
Radiobutton(my_window, text='PARTIAL SCHOLAR', value=2, variable=StudentType, command=on_click).grid(row=9, column=0, sticky='w')
Radiobutton(my_window, text='REGULAR', value=3, variable=StudentType, command=on_click).grid(row=10, column=0, sticky='w')

# FOR OUTPUT TOTAL FEES
Label(my_window, text='TOTAL FEES').grid(row=11, column=0, sticky='w')
Entry(my_window, textvariable=txtTotal, state='disabled').grid(row=11, column=1, sticky='w')

# BUTTONS
Button(my_window, text='COMPUTE', command=on_compute, width=10).grid(row=12, column=0, sticky='w')
Button(my_window, text='CLEAR', command=on_clear, width=10).grid(row=12, column=1, sticky='w')
Button(my_window, text='CLOSE', command=my_window.destroy, width=10).grid(row=12, column=2, sticky='w')

my_window.mainloop()
```
