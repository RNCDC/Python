```python
%run NB2.ipynb
try: 
    score=int(input("Quiz 1 Score: "))
    items=int(input("No. of Items: "))
    if score > items:
        raise Exception("Score should be lower than the No. of Items")
    else:
        q1tg=transg(score,items)
        q1eq=equi(score,items)
        print(" ")
        #OUTPUT FOR QUIZ 1
        print("Quiz 1 TG: %.2f"%q1tg)
        print("Quiz 1 EQ: ",q1eq)
        print(" ")
    score=int(input("Quiz 2 Score: "))
    items=int(input("No. of Items: "))
    if score > items:
        raise Exception("Score should be lower than the No. of Items")
    else:
        q2tg=transg(score,items)
        q2eq=equi(score,items)
        print(" ")
        #OUTPUT FOR QUIZ 2
        print("Quiz 2 TG: %.2f"%q2tg)
        print("Quiz 2 EQ: ",q2eq)
        print(" ")
    avequiz=(q1eq+ q2eq)/2
    quizpercent=((q1eq+ q2eq)/2)*(.35)
    print("Average (EQ): %.2f"%avequiz)
    print("35percent is %.2f"%quizpercent)
    print(" ")
    
    lab1=int(input("Lab 1: "))
    score=lab1
    items=300
    if score > items:
        raise Exception("Score should be lower than the No. of Items")
    else:
        lab2=int(input("Lab 2: "))
        score=(lab1+lab2)
        items=300
        if score > items:
            raise Exception("Score should be lower than the No. of Items")
        else:
            lab3=int(input("Lab 3: "))
            score=(lab1+lab2+lab3)
            items=300
            if score > items:
                raise Exception("Score should be lower than the No. of Items")
            else:
                labeq=equi(score,items)
                print(" ")
                #OUTPUT FOR LAB
                labpercent=labeq*(.35)
                print("Lab Average: %.2f"%labeq)
                print("35percent is %.2f"%labpercent)
                print(" ")
    
    score=int(input("MT Exam Score: "))
    items=int(input("No. of Items: "))
    if score > items:
        raise Exception("Score should be lower than the No. of Items")
    else:
        mtg=transg(score,items)
        meq=equi(score,items)
        print(" ")
        #OUTPUT FOR MIDTERMS
        midpercent=meq*(.30)
        print("Midterm TG: %.2f"%mtg)
        print("Midterm EQ: ",meq)
        print("30percent is %.2f"%midpercent)
        print(" ")
        
    #OUTPUT FOR MIDTERM GRADE    
    mg=quizpercent+labpercent+midpercent
    print("Midterm Grade: %.2f"%mg)
except ValueError: 
    print("Error: Please provide the correct input")
except NameError: 
    print("Error: Please provide the correct input")
except Exception as score:
    print("Error:",score)
```
