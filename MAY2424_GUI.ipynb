```python
#modification of 1 with exception
from tkinter import *
import tkinter.messagebox as tm
my_window = Tk()
my_window.title('Gui Demo')

def on_click():
    try:
        name = txtName.get()
        greet = 'Hello ' + name
        lblName.set(greet)
    except Exception as z:
        tm.showerror('Error: ',z)
    
def on_clear():
    txtName.set(' ')
    lblName.set(' ')
    
def on_close():
    ans = tm.askyesno('Close', 'Are you sure?') 
    #tm is tkinter.messagebox
    #1st parameter title of the window
    #2nd parameter content
    if ans == 1:
        my_window.destroy()
    
#define components
txtName = StringVar()
lblName = StringVar()

Label(my_window, text = 'Name please').grid(row = 0, column = 0)
Entry(my_window, textvariable = txtName, width = 50, border = 3).grid(row = 0, column = 1)
#columnspan para pwede sa gitna ng columns ung ilalagay
Label(my_window, textvariable = lblName).grid(row = 1, column = 0)
Button(my_window, text = 'Click me', command = on_click, width = 10).grid(row = 2, column = 0,columnspan=1)
Button(my_window, text = 'Clear', command = on_clear, width = 10).grid(row = 2, column = 1,columnspan=2)
Button(my_window, text = 'Close', command = on_close, width = 10).grid(row = 2, column = 2,columnspan=3)

my_window.mainloop()
```

```python
#2 TG
from tkinter import *
import tkinter.messagebox as tm
my_window = Tk()
my_window.title('Gui Demo')

def on_click():
    try:
        score = txtScore.get()
        items = txtItems.get()
        tg = int(score) * 50 / int(items) + 50
        txtTG.set(tg)
    except Exception as z:
        tm.showerror('Error: ',z)
    
def on_clear():
    txtScore.set(' ')
    txtItems.set(' ')
    txtTG.set(' ')
    
def on_close():
    ans = tm.askyesno('Close', 'Are you sure?') 
    #tm is tkinter.messagebox
    #1st parameter title of the window
    #2nd parameter content
    if ans == 1:
        my_window.destroy()
    
#define components
txtScore = StringVar()
txtItems = StringVar()
txtTG = StringVar()
lblName = StringVar()

Label(my_window, text = 'Score: ').grid(row = 0, column = 0)
Entry(my_window, textvariable = txtScore, width = 20, border = 3).grid(row = 0, column = 1)
Label(my_window, text = 'Number of Items: ').grid(row = 1, column = 0)
Entry(my_window, textvariable = txtItems, width = 20, border = 3).grid(row = 1, column = 1)
Label(my_window, text = 'Transmuted Grade: ').grid(row = 2, column = 0)
Entry(my_window, textvariable = txtTG, width = 20, border = 3, state = 'disabled').grid(row = 2, column = 1)
Button(my_window, text = 'Compute', command = on_click, width = 10).grid(row = 3, column = 0)
Button(my_window, text = 'Clear', command = on_clear, width = 10).grid(row = 3, column = 1)
Button(my_window, text = 'Close', command = on_close, width = 10).grid(row = 3, column = 2)

my_window.mainloop()
```

```python
#ACTIVITY
from tkinter import *
import tkinter.messagebox as tm
my_window = Tk()
my_window.title('Final grade')

#methods
def on_click():
    try:
        mid = txtMid.get()
        fin = txtFin.get()
        ave = (float(mid) + float(fin)) / 2
        txtAve.set('%.2f' %ave)
        
        if ave >= 3.13:
            txtRemarks.set('Failed')
        else:
            txtRemarks.set('Passed')
            
    except Exception as z:
        tm.showerror('Error: ',z)
    
def on_clear():
    txtMid.set(' ')
    txtFin.set(' ')
    txtAve.set(' ')
    txtRemarks.set(' ')
    
def on_close():
    ans = tm.askyesno('Close', 'Are you sure?') 
    #tm is tkinter.messagebox
    #1st parameter title of the window
    #2nd parameter content
    if ans == 1:
        my_window.destroy()
    
#define components
txtMid = StringVar()
txtFin = StringVar()
txtAve = StringVar()
txtRemarks = StringVar()

Label(my_window, text = 'Midterm: ').grid(row = 0, column = 0)
Entry(my_window, textvariable = txtMid, width = 20, border = 3).grid(row = 0, column = 1)
Label(my_window, text = 'Finals: ').grid(row = 1, column = 0)
Entry(my_window, textvariable = txtFin, width = 20, border = 3).grid(row = 1, column = 1)
Label(my_window, text = 'Average: ').grid(row = 2, column = 0)
Entry(my_window, textvariable = txtAve, width = 20, border = 3, state = 'disabled').grid(row = 2, column = 1)
Label(my_window, text = 'Remarks: ').grid(row = 3, column = 0)
Entry(my_window, textvariable = txtRemarks, width = 20, border = 3, state = 'disabled').grid(row = 3, column = 1)
Button(my_window, text = 'Compute', command = on_click, width = 10).grid(row = 4, column = 0)
Button(my_window, text = 'Clear', command = on_clear, width = 10).grid(row = 4, column = 1)
Button(my_window, text = 'Close', command = on_close, width = 10).grid(row = 4, column = 2)

my_window.mainloop()
```
