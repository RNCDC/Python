```python
#5 final
import tkinter as tk

class Circle:
    def __init__(self):
        self.radius = 0.0
        self.area = 0.0
        self.diameter = 0.0
    
    def setRadius(self, radius):
        self.radius = radius
        self.computeDiameter()
        self.computeArea()
    
    def computeDiameter(self):
        self.diameter = 2 * self.radius
    
    def computeArea(self):
        self.area = 3.14 * (self.radius ** 2) #** exponent

class TestCircle:
    def __init__(self, root):
        self.root = root
        self.root.title("Circle Properties")
        
        self.circles = {
            'Small': {'circle': Circle()},
            'Large': {'circle': Circle()},
            'Default': {'circle': Circle()}
        }
        
        row = 0
        for label in self.circles:
            self.create_circle_section(label, row)
            row += 1
        
        self.circles['Default']['circle'].setRadius(1)

    def create_circle_section(self, label, row):
        tk.Label(self.root, text=f"{label} Circle").grid(row=row, column=0)
        
        radius_entry = tk.Entry(self.root)
        radius_entry.grid(row=row, column=1)
        
        if label == 'Default':
            radius_entry.insert(0, '1')
            radius_entry.config(state="disabled")
        
        compute_button = tk.Button(self.root, text="Compute", command=lambda label=label: self.compute_circle(label))
        compute_button.grid(row=row, column=2)
        
        area_label = tk.Label(self.root, text="Area:")
        area_label.grid(row=row, column=3)
        
        diameter_label = tk.Label(self.root, text="Diameter:")
        diameter_label.grid(row=row, column=4)
        
        self.circles[label].update({'entry': radius_entry, 'area_label': area_label, 'diameter_label': diameter_label})

    #for exception
    def compute_circle(self, label):
        try:
            radius = float(self.circles[label]['entry'].get())
            self.circles[label]['circle'].setRadius(radius)
            self.update_circle_labels(label)
        except ValueError:
            self.circles[label]['area_label'].config(text="Invalid input!")
            self.circles[label]['diameter_label'].config(text="Invalid input!")

    def update_circle_labels(self, label):
        circle = self.circles[label]['circle']
        self.circles[label]['area_label'].config(text=f"Area: {circle.area:.2f}")
        self.circles[label]['diameter_label'].config(text=f"Diameter: {circle.diameter:.2f}")

    def compute_circle(self, label):
        try:
            radius = float(self.circles[label]['entry'].get())
            if radius <= 0:
                raise ValueError("Radius must be greater than zero.")
            
            self.circles[label]['circle'].setRadius(radius)
            self.update_circle_labels(label)
        except ValueError as e:
            self.circles[label]['area_label'].config(text=str(e))
            self.circles[label]['diameter_label'].config(text=str(e))
        

if __name__ == "__main__":
    root = tk.Tk()
    app = TestCircle(root)
    root.mainloop()
```

```python
#5
import tkinter as tk

class Circle:
    def __init__(self):
        self.radius = 0.0
        self.area = 0.0
        self.diameter = 0.0
    
    def setRadius(self, radius):
        self.radius = radius
    
    def getRadius(self):
        return self.radius
    
    def computeDiameter(self):
        self.diameter = 2 * self.radius
    
    def computeArea(self):
        self.area = 3.14 * (self.radius ** 2)

class TestCircle:
    def __init__(self, root):
        self.root = root
        self.root.title("Circle Properties")
        
        self.small_circle = Circle()
        self.large_circle = Circle()
        self.default_circle = Circle()
        
        # Labels
        self.small_label = tk.Label(root, text="Small Circle")
        self.small_label.grid(row=0, column=0)
        self.large_label = tk.Label(root, text="Large Circle")
        self.large_label.grid(row=0, column=2)
        self.default_label = tk.Label(root, text="Default Circle")
        self.default_label.grid(row=0, column=4)
        
        # Radius input fields
        self.small_radius_entry = tk.Entry(root)
        self.small_radius_entry.grid(row=1, column=0)
        self.large_radius_entry = tk.Entry(root)
        self.large_radius_entry.grid(row=1, column=2)
        self.default_radius_entry = tk.Entry(root)
        self.default_radius_entry.grid(row=1, column=4)
        self.default_radius_entry.insert(0, '1')  # Set default value
        
        # Area and Diameter labels (initially empty)
        self.small_area_label = tk.Label(root, text="Area:")
        self.small_area_label.grid(row=2, column=0)
        self.small_diameter_label = tk.Label(root, text="Diameter:")
        self.small_diameter_label.grid(row=3, column=0)
        
        self.large_area_label = tk.Label(root, text="Area:")
        self.large_area_label.grid(row=2, column=2)
        self.large_diameter_label = tk.Label(root, text="Diameter:")
        self.large_diameter_label.grid(row=3, column=2)
        
        self.default_area_label = tk.Label(root, text="Area:")
        self.default_area_label.grid(row=2, column=4)
        self.default_diameter_label = tk.Label(root, text="Diameter:")
        self.default_diameter_label.grid(row=3, column=4)
        
        # Compute buttons
        self.small_compute_button = tk.Button(root, text="Compute", command=self.compute_small_circle)
        self.small_compute_button.grid(row=4, column=0)
        self.large_compute_button = tk.Button(root, text="Compute", command=self.compute_large_circle)
        self.large_compute_button.grid(row=4, column=2)
        self.default_compute_button = tk.Button(root, text="Compute", command=self.compute_default_circle)
        self.default_compute_button.grid(row=4, column=4)
        
        # Disable non-input fields initially
        self.disable_fields()

    def compute_small_circle(self):
        try:
            radius = float(self.small_radius_entry.get())
            self.small_circle.setRadius(radius)
            self.small_circle.computeDiameter()
            self.small_circle.computeArea()
            self.small_area_label.config(text=f"Area: {self.small_circle.area:.2f}")
            self.small_diameter_label.config(text=f"Diameter: {self.small_circle.diameter:.2f}")
        except ValueError:
            self.small_area_label.config(text="Invalid input!")
            self.small_diameter_label.config(text="Invalid input!")
    
    def compute_large_circle(self):
        try:
            radius = float(self.large_radius_entry.get())
            self.large_circle.setRadius(radius)
            self.large_circle.computeDiameter()
            self.large_circle.computeArea()
            self.large_area_label.config(text=f"Area: {self.large_circle.area:.2f}")
            self.large_diameter_label.config(text=f"Diameter: {self.large_circle.diameter:.2f}")
        except ValueError:
            self.large_area_label.config(text="Invalid input!")
            self.large_diameter_label.config(text="Invalid input!")
    
    def compute_default_circle(self):
        try:
            radius = float(self.default_radius_entry.get())
            self.default_circle.setRadius(radius)
            self.default_circle.computeDiameter()
            self.default_circle.computeArea()
            self.default_area_label.config(text=f"Area: {self.default_circle.area:.2f}")
            self.default_diameter_label.config(text=f"Diameter: {self.default_circle.diameter:.2f}")
        except ValueError:
            self.default_area_label.config(text="Invalid input!")
            self.default_diameter_label.config(text="Invalid input!")
    
    def disable_fields(self):
        # Disable non-input fields initially
        self.small_area_label.config(state="disabled")
        self.small_diameter_label.config(state="disabled")
        self.large_area_label.config(state="disabled")
        self.large_diameter_label.config(state="disabled")
        self.default_area_label.config(state="disabled")
        self.default_diameter_label.config(state="disabled")
        self.default_radius_entry.config(state="disabled")  # Disable default circle entry

if __name__ == "__main__":
    root = tk.Tk()
    app = TestCircle(root)
    root.mainloop()
```
